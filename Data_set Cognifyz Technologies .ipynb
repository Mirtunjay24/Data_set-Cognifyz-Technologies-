```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram of Age
sns.histplot(df['age'], bins=10, kde=True)
plt.title("Distribution of Age")
plt.show()

# Relationship between Age and Salary
sns.scatterplot(x='fixed_deposits', y='equity_market', data=df)
plt.title("fixed vs equaity")
plt.show()
```

```python

sns.histplot(df['age'], bins=10, kde=True)
plt.title("Age Distribution of Investors")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
df['investment_avenues'].value_counts().plot.pie(autopct='%1.1f%%', figsize=(6,6))
plt.title("Preference for Investment Avenues")
plt.ylabel("")  # Hide y-label
plt.show()
```

```python
sns.boxplot(x='gender', y='mutual_funds', data=df)
plt.title("Mutual Fund Investments by Gender")
plt.xlabel("Gender")
plt.ylabel("Investment Level")
plt.show()
```

```python
sns.countplot(x='invest_monitor', data=df, order=df['invest_monitor'].value_counts().index)
plt.title("Investment Monitoring Frequency")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.countplot(y='factor', data=df, order=df['factor'].value_counts().index)
plt.title("Factors Influencing Investment Decisions")
plt.xlabel("Count")
plt.ylabel("Factor")
plt.show()
```

```python
sns.boxplot(x=pd.cut(df['age'], bins=[18, 25, 35, 50, 70], labels=['18-25', '26-35', '36-50', '51-70']),
            y='mutual_funds', data=df)
plt.title("Mutual Fund Investment Across Age Groups")
plt.xlabel("Age Group")
plt.ylabel("Mutual Fund Investment Level")
plt.show()
```
